```python
import openpyxl
import pandas as pd
import os
```

```python
excelname='raw_data'
sheetname='data'
wb=openpyxl.load_workbook(str(excelname)+'.xlsx')
ws=wb[str(sheetname)]
```

```python
dataset=[]
i=1
while(ws.cell(row=i,column=1).value!=None):
    row=[]
    j=1
    while(ws.cell(row=i,column=j).value!=None):
        row.append(ws.cell(row=i,column=j).value)
        j+=1
    dataset.append(row)
    i+=1
```

```python
trade=[[0 for i in range(16)]for j in range(len(dataset))]
trade[0][0]='date'
trade[0][1]='open'
trade[0][2]='high'
trade[0][3]='low'
trade[0][4]='close'
trade[0][5]='Aroon_Up'
trade[0][6]='Aroon_Down'
trade[0][7]='Aroon_Oscillator'
trade[0][8]='buy'
trade[0][9]='hold'
trade[0][10]='sell'
trade[0][11]='number'
trade[0][12]='value'
trade[0][13]='period_return'
trade[0][14]='holding_period'
trade[0][15]='annual_return'
for i in range(1,len(trade)):
    trade[i][0]=dataset[i][0]
    trade[i][1]=dataset[i][1]
    trade[i][2]=dataset[i][2]
    trade[i][3]=dataset[i][3]
    trade[i][4]=float(dataset[i][4])
```

```python
pd.DataFrame(trade).head(3)
```

```python
Aroon_index = 25
# Aroon Oscillator(Aroon_Up & Aroon_Down)
for i in range(Aroon_index+1,len(trade)):
# for i in range(Aroon_index+1, 27):
    period_high = -10000
    period_high_index = 0
    period_low  = 10000
    period_low_index = 0
    for j in range(0, Aroon_index+1):
        if(trade[i-j][2]>period_high):
            period_high = trade[i-j][2]
            period_high_index = 0
        if(period_high > trade[i-j][2]):
            period_high_index = period_high_index +1
        if(trade[i-j][3]<period_low):
            period_low = trade[i-j][3]
            period_low_index = 0
        if(period_low < trade[i-j][3]):
            period_low_index = period_low_index + 1
    period_high_index = Aroon_index - period_high_index
    period_low_index  = Aroon_index - period_low_index
    trade[i][5]= ((Aroon_index-period_high_index)/Aroon_index)*100
    trade[i][6]= ((Aroon_index-period_low_index)/Aroon_index)*100
    trade[i][7]=trade[i][5]-trade[i][6]
```

```python
#import datetime             
import datetime
from datetime import date
for i in range(1,len(trade)):
    trade[i][0]=datetime.datetime.strptime(trade[i][0],'%Y/%m/%d')
#holding_period
first=0
end=0
for i in range(1,len(trade)):
    if(trade[i][8]==1):
        trade[i][14]=1
        first=trade[i][0]
    elif((trade[i][9]>0)or(trade[i][10]==1)and(trade[i][8]!=1)):
        end=trade[i][0]
        trade[i][14]=(end-first).days+1
#annual_return (holding_period)
for i in range(1,len(trade)):
    if(trade[i][14]!=0):
        trade[i][15]=(1+trade[i][13])**(365/trade[i][14])-1
#annual_return (trading)
annual_return=1
trading=0
for i in range(1,len(trade)):
    if(trade[i][10]==1):
        annual_return*=(1+trade[i][13])
        trading+=1

final_return=annual_return**(1/trading)-1
```

```python
def hyper(dataset, Aroon_index):
    trade=[[0 for i in range(16)]for j in range(len(dataset))]
    trade[0][0]='date'
    trade[0][1]='open'
    trade[0][2]='high'
    trade[0][3]='low'
    trade[0][4]='close'
    trade[0][5]='Aroon_Up'
    trade[0][6]='Aroon_Down'
    trade[0][7]='Aroon_Oscillator'
    trade[0][8]='buy'
    trade[0][9]='hold'
    trade[0][10]='sell'
    trade[0][11]='number'
    trade[0][12]='value'
    trade[0][13]='period_return'
    trade[0][14]='holding_period'
    trade[0][15]='annual_return'
    for i in range(1,len(trade)):
        trade[i][0]=dataset[i][0]
        trade[i][1]=dataset[i][1]
        trade[i][2]=dataset[i][2]
        trade[i][3]=dataset[i][3]
        trade[i][4]=float(dataset[i][4])
#     Aroon_index = 25
    # Aroon Oscillator(Aroon_Up & Aroon_Down)
    for i in range(Aroon_index+1,len(trade)):
    # for i in range(Aroon_index+1, 27):
        period_high = -10000
        period_high_index = 0
        period_low  = 10000
        period_low_index = 0
        for j in range(0, Aroon_index+1):
            if(trade[i-j][2]>period_high):
                period_high = trade[i-j][2]
                period_high_index = 0
            if(period_high > trade[i-j][2]):
                period_high_index = period_high_index +1
            if(trade[i-j][3]<period_low):
                period_low = trade[i-j][3]
                period_low_index = 0
            if(period_low < trade[i-j][3]):
                period_low_index = period_low_index + 1
        period_high_index = Aroon_index - period_high_index
        period_low_index  = Aroon_index - period_low_index
        trade[i][5]= ((Aroon_index-period_high_index)/Aroon_index)*100
        trade[i][6]= ((Aroon_index-period_low_index)/Aroon_index)*100
        trade[i][7]=trade[i][5]-trade[i][6]
        
    for i in range(Aroon_index,len(trade)):
        if(trade[i][7]>0)and(trade[i-1][9]==0):
            trade[i][8]=1
            trade[i][9]=1
            trade[i][10]=0
        elif(trade[i][7]<0)and(trade[i-1][9]==1):
            trade[i][8]=0
            trade[i][9]=0
            trade[i][10]=1
        elif(trade[i-1][9]>0):
            trade[i][8]=0
            trade[i][9]=1
            trade[i][10]=0
            
    initial=10000
    for i in range(25,len(trade)):
        if(trade[i][8]==1)and(trade[i][9]==1)and(trade[i][10]==0):
            trade[i][11]=initial/trade[i][4]
            trade[i][12]=trade[i][11]*trade[i][4]
            trade[i][13]=round((trade[i][11]*trade[i][4]-initial)/initial,4)
        elif(trade[i][8]==0)and(trade[i][9]==0)and(trade[i][10]==1):
            trade[i][11]=0
            trade[i][12]=trade[i-1][11]*trade[i][4]
            trade[i][13]=round((trade[i-1][11]*trade[i][4]-initial)/initial,4)
        elif(trade[i][8]==0)and(trade[i][9]==1)and(trade[i][10]==0):
            trade[i][11]=trade[i-1][11]
            trade[i][12]=trade[i][11]*trade[i][4]
            trade[i][13]=round((trade[i][11]*trade[i][4]-initial)/initial,4)
    #import datetime             
    import datetime
    from datetime import date
    for i in range(1,len(trade)):
        trade[i][0]=datetime.datetime.strptime(trade[i][0],'%Y/%m/%d')
    #holding_period
    first=0
    end=0
    for i in range(1,len(trade)):
        if(trade[i][8]==1):
            trade[i][14]=1
            first=trade[i][0]
        elif((trade[i][9]>0)or(trade[i][10]==1)and(trade[i][8]!=1)):
            end=trade[i][0]
            trade[i][14]=(end-first).days+1
    #annual_return (holding_period)
    for i in range(1,len(trade)):
        if(trade[i][14]!=0):
            trade[i][15]=(1+trade[i][13])**(365/trade[i][14])-1
    #annual_return (trading)
    annual_return=1
    trading=0
    for i in range(1,len(trade)):
        if(trade[i][10]==1):
            annual_return*=(1+trade[i][13])
            trading+=1

    final_return=annual_return**(1/trading)-1
    return final_return
```
